### Answer 1

The process of Understanding involves three steps, where each one is dependant on the previous one:

1. **Perceiving**: It concerns the act of being able to read a chart. The viewer attempts to decode the representation of the data (shapes, sizes, colours) as displayed and convert them into perceived values (estimates of quantities and their relationships to other values).
Usual questions are:

 - Where are the largest, middle-sized and smallest values?
 - What proportion of the total does that value hold?
 - How do these values compare in ranking terms?
 - To which other values does this have a connected relationship?


2. **Interpreting**: After Having read the chart, now the viewer tries to convert these perceived values into some form of meaning. Usual questions are:

 - Is it good to be big or better to be small?
 - What does it mean to go up or down?
 - Is that relationship meaningful or insignificant?
 - Is the decline of that category especially surprising?
 - This step depends on viewer's pre-existing knowledge about the subject and his capacity to utilise that knowledge to frame the implications of what has been read. If the viewer do not possess that knowledge, it may be up to the creator to address that lackness. to achieve this, the visualizer can use captions, headlines, colours and other annotated devices.


3. **Comprehending**: it involves reasoning the consequence of the two previous steps (Perceiving and Interpreting) to arrive at a personal reflection of what this visualisation means to the viewer and how this make a difference with respect to what he already knew about the subject. Usual questions are:

 - Why is this relevant? What wants or needs does it serves?
 - Has it confirmed what I knew or suspected beforhand, or enlightened me with some new knowledge?
 - Has this experience impacted me in an emotional way or left me feeling indifferent as a consequence?
 - Does the new understanding I have acquired do push me to take action (make a decision or change my behaviour) or it just enhance my knowledge about the subject?

for the categorical one:

**Bar chart**: 

```python
img = mpimg.imread('jan_exam_ex2_1.jpg')

fig, axes = plt.subplots(1, 1, figsize = (14, 8))

plt.xticks([])
plt.yticks([])

for i in ['top', 'left', 'bottom', 'right']:
    plt.gca().spines[i].set_visible(False)

plt.imshow(img)
plt.show()
```

In this kind of chart, each bar represent a categorical variable. In this case each one is associated to a given actor. While looking at it you can also have an idea about maximum and minimum values. In this case is pretty easy since each category is ranked based on its y value. Also neighbouring comparison is easy and allows to identify any kind of relationship. Usually in bar charts tick marks and gridline helps the reader to be accurate while reading quantitative values. As a general rule the origin value should always be at 0. As a good practice, bars should always have the exact size of the quantity they are representing, this means that using figures instead of bars could distort the quantity perception and it is better to avoid, even if it may be seen as appealing (this of course relate also to trustworthiness). Apart from this specific case (vertical bar chart), also an horizontal one can be used. And it is quite useful when the bar chart start to be populated by more than some categories. Differently from histograms, bar charts always have gaps between each bar, this is to avoid giving a false idea of continuity between each value.

```python
img = mpimg.imread('jan_exam_ex2_2.jpg')

fig, axes = plt.subplots(1, 1, figsize = (14, 8))

plt.xticks([])
plt.yticks([])

for i in ['top', 'left', 'bottom', 'right']:
    plt.gca().spines[i].set_visible(False)

plt.imshow(img)
plt.show()
```

Scatterplots display the relationships between two different quantitative variables. These are useful in order to explore and find potential relationships between categories (x and y). Sometimes color is used in order to display different categories for those same variables. When one or both the variables have a small range, the chart may become to cluttered, making it very difficult to search for relationships.
About the relationships one can see, it is useful looking for any diagonal trends or, on the contrary, a flat line (no correlation at all). Sometimes, when all the points stay more or less very close eachothers, the outliers start to become interesting. Usually, also looking at eventual clusters between the same categories is relevant. As for bar charts, annotations such as gridlines may help make precise value estimates. Also a fit line helps understanding better the relationship. Labeling certain values may be useful, but just the most noticeable ones (lowest, highest, interesting for some reasons), unless interactivity is provided.
About categories, it is suggested to use very different colors in order to be able to distinguish them clearly.
Differently from bar plots, scatterplots do not need to have a zero origin.
As a general hint, if one variable is said to be dependant to the other, the first one usually goes to the y axis. Data transformation has to be stated along the axis (e.g., log transformation), so to avoid misunderstanding and wrong conclusions by the reader.

A choropleth map displays quantitative values for different spatial regions on a map. Each one is represented by a polygonal area, usually this happens through a geojson file that contains such information. Each region is color-coded to represent a value based on a scale, where colors usually go from lighter (small values) to darker ones (high values).

When you are in front of this kind of visualization you are going to look for light, medium and dark colors, associated to small, medium and large values. In this way is easy to make comparison between different geographical areas. Of course, the size of the geographical area has no meaning at all wrt values, since is just a representation of the actual underlying region.
Labels may be useful if in a limited number.
The color scale legend should be as close as possible to the map, geographical borders should be visible, but not that much to dominate the attention of the viewer.

## Question 3

Describe the PCA algorithm, highlighting its pros and cons; find a
case study where PCA is more appropriate than MDS.

### Answer 3

Principal Component Analysis is a dimensionality reduction algorithm. The main steps of PCA are:

- standardization: to standardize the range of the continuous variables so that each of them contribute equally to the analysis. A common way to do that is by computing Z-score: $z = \frac{\text{data } - \text{ mean}}{\text{standard deviation}}$; this way all the data will have mean zero and standard deviation one.

- covariance matrix: its computation is useful in order to understand how variables vary from the mean with respect to each other. In other words, to understand if any data is correlated: if that is the case, there is going to be some information redundancy. So, the covariance matrix allows to see how variables covariate (they could be positive or negatively correlated).

- extract eigenvalues and eigenvectors: this is the step in which principal components are computed. PCs are new variables constructed as linear combinations of the starting variables, in such a way that they are uncorrelated and carrying the highest possible variance. PCA builds up new variables (PCs) where the first ones are those carrying most of the variance, so that each subsequent PC carries less variance than the previous one. So, the aim is to reduce dimensionality while keeping most of the explained variance. The first PC is found maximizing the variance (i.e. finding the eigenvector associated to the largest eigenvalue). The following ones are computed in the same way, but they have to be uncorrelated (orthogonal) to the previous ones.

- data 'recasting': finding PCs does not change the actual data, which, in fact, it is still represented in terms of the original axes. So, in this step, data has to be reoriented from the original axes to the one represented by the PCs. This is done by multiplying the transpose of the original data by the transpose of the 'feature vector' (which is the matrix formed by the previously selected eigenvectors).

In practice, PCA is computed by three possible methods:

- Covariance, used when data have similar scales;
- Correlation, used when data have different scales;
- Singular Value Decomposition, more general purpose.


PROS:

- It is good at extracting signal from noise
- Computationally inexpensive (time wise)
- Good for linear data


CONS:

- Not appropriate for non-linear data
- Variables are less interpretable after transformation
- Data standardization is required
- Information loss, even if PCA tries to maximize the variance, the number of PCs has to be selected with care.

An example where we can prefer PCA over MDS is the one of the USArrests data set. For each of the 50 states in the United States, the data set contains the number of arrestsper 100000 residents for each of three crimes: Assault, Murder, and Rape. We also record UrbanPop (the percent of the population in each state living in urban areas). The results of the PCA on this dataset show that the first loading vector places approximately equal weight on Assault, Murder, and Rape, with much less weight on UrbanPop. Hence this component roughly corresponds to a measure of overall rates of serious crimes. The second loading vector places most of its weight on UrbanPop and much less weight on the other three features. Hence, this component roughly corresponds to the level of urbanization of the state. Overall, we see that the crime-related variables (Murder, Assault, and Rape) are located close to each other, and that the UrbanPop variable is far from the other three. This indicates that the crime-related variables are correlated with each otherstates with high murder rates tend to have high assault and rape ratesand that the UrbanPop variable is less correlated with the other three.

I have also found this:

- The Duff Moisture Code (DMC) represents fuel moisture of decomposed organic material underneath the litter. System designers suggest that it is represents moisture conditions for the equivalent of 15-day (or 360 hr) timelag fuels. It is unitless and open ended. It may provide insight to live fuel moisture stress.

- The Drought Code (DC), much like the Keetch-Byrum Drought Index, represents drying deep into the soil. It approximates moisture conditions for the equivalent of 53-day (1272 hour) timelag fuels. It is unitless, with a maximum value of 1000. Extreme drought conditions have produced DC values near 800.

So, this kind of correlation is pretty much trivial, since are two similar measures of the same thing

```python
y = df.groupby(by = 'month').count().iloc[:, 0]
y.values
```

```python
values = []

x = list(y.index)
y = list(y.values)

for i, j in zip(x, y):
    values.append((i, j))
    
values = sorted(values, key = lambda x : x[1])

```

```python
plt.bar(x = [i[0] for i in values], height = [i[1] for i in values])

plt.title('Fire events for month')
plt.xlabel('month')
plt.ylabel('# of fires')

plt.show()
```

```python
# as I was expecting, the most hot months are also the ones with more fires
```

```python
a = df.groupby(['temp', 'RH']).X.count()
temp = [i[0] for i in a.index]
humidity = [i[1] for i in a.index]
```

```python
new = pd.DataFrame(data = [temp, humidity, a.values]).transpose()
new.columns = ['temp', 'humidity', 'count']
```

```python
sns.scatterplot(x = new.temp, y = new.humidity, size = [int(i) for i in new['count']], hue = [int(i) for i in new['count']],
               palette = sns.color_palette('Reds', 4), alpha = 0.8, zorder = 3)
sns.regplot(x = new.temp, y = new.humidity, scatter = False, ci = False, order = 1, color = 'red', line_kws = {'linewidth' : 1, 'linestyle' : '--'})

plt.title('Number of fires for a given humidity and temperature')

plt.legend(title = 'Counts: ')

plt.xlim(0, 35)
plt.ylim(0, 110)


plt.show()
```

As you can see, much more fire events are after 15 degreee and when the humidity is relatively low (lower than 50).

A better visualisation maybe a contourplot, since it "inglobate" near points and allows us to see the density of the events

```python
#it is weigheted based on the number of counts for that specific pair.

fig, ax = plt.subplots(1, 1, figsize = (14, 8))

sns.kdeplot(x = new.temp, y = new.humidity, cmap = "Reds", shade = True, bw_adjust = 0.5, weights = new['count'], 
            zorder = 0)

plt.title('2d density of fire events based on humidity and temp')

ax.yaxis.set_ticks_position('none'), ax.xaxis.set_ticks_position('none')

ax.grid(which = 'major', axis = 'both', zorder = 2, color = 'k', linewidth = 0.2, linestyle = '--')

plt.xlabel('Temp')
plt.ylabel('Humidity')

for i in ['top', 'left', 'bottom', 'right']:
    if i == 'left' or i == 'bottom':
        plt.gca().spines[i].set_alpha(0.2)
    else:
        plt.gca().spines[i].set_visible(False)
    

plt.show()
```

### Answer part 1

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('forestfires.csv')
```

```python
df = df[(df.month == 'jan') | (df.month == 'jul')]
```

```python
df.reset_index(inplace = True)
```

```python
lst = []

for i, j in zip(df.X, df.Y):
    mn = df[(df.X == i) & (df.Y == j)].temp.mean()
    if (i, j, mn) not in lst:
        lst.append((i, j, mn))
```

```python
mat = np.zeros((10, 10))

for i in lst:
    x = i[0]
    y = i[1]
    mn = i[2]
    mat[x, y] = mn
```

```python
#reversed Y values

for i in range(len(mat)):
    mat[i] = mat[i][::-1]
```

```python
fig, ax = plt.subplots(1, 1, figsize = (8, 8))

sns.heatmap(mat[::-1], cmap = 'Reds', xticklabels = np.arange(0, 10), yticklabels = np.arange(9, 0, -1),
            linewidths = 0.01)

plt.title('Avg. Temperature (January and July) on positions')

plt.xlabel('X')
plt.ylabel('Y')

ax.yaxis.set_ticks_position('none'), ax.xaxis.set_ticks_position('none')


ax.set_aspect('equal')

plt.show()
```
